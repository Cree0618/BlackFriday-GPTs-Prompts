```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Load your data
data = pd.read_csv('/content/data_analyze_modified.csv')

# Clean and preprocess your data
data = data.drop(columns=['Unnamed: 3', 'Unnamed: 4'])
data['date'] = data['date'].str.strip(',').apply(lambda x: pd.to_datetime(x, errors='coerce'))
data = data.dropna(subset=['date']).reset_index(drop=True)
data['ma_7'] = data['ma_7'].fillna(method='ffill')
data['ma_7'] = data['ma_7'].fillna(data['ma_7'].mean())

# Feature engineering
data['day'] = data['date'].dt.day
data['month'] = data['date'].dt.month
data['year'] = data['date'].dt.year

# Prepare features and target variable
X = data[['date', 'ma_7', 'day', 'month', 'year']]
y = data['spot_price']

# Convert 'date' to numerical format (ordinal)
X['date'] = X['date'].map(pd.Timestamp.toordinal)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the standardized data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Define different optimizers to compare
optimizers = ['adam', 'sgd', 'lbfgs']

# Dictionary to store the results
results = {}

for optimizer in optimizers:
    # Initialize and train the neural network model with the current optimizer
    nn_model = MLPRegressor(hidden_layer_sizes=(100, 100), max_iter=500, solver=optimizer, random_state=42)
    nn_model.fit(X_train, y_train)

    # Predict on the test set
    y_pred_nn = nn_model.predict(X_test)

    # Evaluate the neural network model
    mae_nn = mean_absolute_error(y_test, y_pred_nn)
    mse_nn = mean_squared_error(y_test, y_pred_nn)
    r2_nn = r2_score(y_test, y_pred_nn)

    # Store the results
    results[optimizer] = {'MAE': mae_nn, 'MSE': mse_nn, 'R2': r2_nn}

    # Print the results for the current optimizer
    print(f"Optimizer: {optimizer}, MAE: {mae_nn}, MSE: {mse_nn}, R2: {r2_nn}")

# Visualize the results
plt.figure(figsize=(12, 6))
for optimizer, result in results.items():
    plt.plot(y_test.values, label=f'Actual - {optimizer}', color='blue')
    plt.plot(y_pred_nn, label=f'Predicted - {optimizer}', color='red')
    plt.title(f'Actual vs Predicted Spot Prices ({optimizer} Optimizer)')
    plt.xlabel('Sample Index')
    plt.ylabel('Spot Price')
    plt.legend()
    plt.show()

# Print the overall comparison
print("Overall comparison of different optimizers:")
for optimizer, result in results.items():
    print(f"Optimizer: {optimizer}, MAE: {result['MAE']}, MSE: {result['MSE']}, R2: {result['R2']}")
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, GridSearchCV
import matplotlib.pyplot as plt

# Load your data
data = pd.read_csv('/content/data_analyze_modified.csv')

# Clean and preprocess your data
data = data.drop(columns=['Unnamed: 3', 'Unnamed: 4'])
data['date'] = data['date'].str.strip(',').apply(lambda x: pd.to_datetime(x, errors='coerce'))
data = data.dropna(subset=['date']).reset_index(drop=True)
data['ma_7'] = data['ma_7'].fillna(method='ffill')
data['ma_7'] = data['ma_7'].fillna(data['ma_7'].mean())

# Feature engineering
data['day'] = data['date'].dt.day
data['month'] = data['date'].dt.month
data['year'] = data['date'].dt.year

# Prepare features and target variable
X = data[['date', 'ma_7', 'day', 'month', 'year']]
y = data['spot_price']

# Convert 'date' to numerical format (ordinal)
X['date'] = X['date'].map(pd.Timestamp.toordinal)

# Apply MinMax scaling to features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split the standardized data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'hidden_layer_sizes': [(50, 50), (100, 100), (100, 100, 100)],
    'activation': ['relu', 'tanh', 'logistic'],
    'alpha': [0.0001, 0.001, 0.01],
    'max_iter': [1000, 2000, 3000]  # Increased max_iter
}

# Initialize the model with lbfgs solver
nn_model = MLPRegressor(solver='lbfgs', random_state=42)

# Perform grid search without parallel processing
grid_search = GridSearchCV(estimator=nn_model, param_grid=param_grid, cv=3, n_jobs=1, verbose=2)
grid_search.fit(X_train, y_train)

# Best model from grid search
best_nn_model = grid_search.best_estimator_

# Predict on the test set with the best model
y_pred_nn = best_nn_model.predict(X_test)

# Evaluate the neural network model
mae_nn = mean_absolute_error(y_test, y_pred_nn)
mse_nn = mean_squared_error(y_test, y_pred_nn)
r2_nn = r2_score(y_test, y_pred_nn)

print(f"Best parameters found: {grid_search.best_params_}")
print(f"MAE: {mae_nn}, MSE: {mse_nn}, R2: {r2_nn}")

# Visualize the results
plt.figure(figsize=(12, 6))
plt.plot(y_test.values, label='Actual', color='blue')
plt.plot(y_pred_nn, label='Predicted', color='red')
plt.title('Actual vs Predicted Spot Prices (Optimized Neural Network)')
plt.xlabel('Sample Index')
plt.ylabel('Spot Price')
plt.legend()
plt.show()
```

```python
#Time Series Forecasting with LSTM (Using Keras)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import numpy as np

# Prepare the data for LSTM
def create_sequences(data, seq_length):
    xs, ys = [], []
    for i in range(len(data)-seq_length):
        x = data[i:i+seq_length]
        y = data[i+seq_length]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

seq_length = 10
X_seq, y_seq = create_sequences(X_scaled, seq_length)

# Split into train and test sets
X_train_seq, X_test_seq, y_train_seq, y_test_seq = train_test_split(X_seq, y_seq, test_size=0.2, random_state=42)

# Build the LSTM model
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(seq_length, X_scaled.shape[1])))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(X_train_seq, y_train_seq, epochs=200, validation_split=0.2, verbose=2)

# Make predictions
y_pred_seq = model.predict(X_test_seq)

# Evaluate the model
mae_lstm = mean_absolute_error(y_test_seq, y_pred_seq)
mse_lstm = mean_squared_error(y_test_seq, y_pred_seq)
r2_lstm = r2_score(y_test_seq, y_pred_seq)

print(f"LSTM MAE: {mae_lstm}, MSE: {mse_lstm}, R2: {r2_lstm}")
```
